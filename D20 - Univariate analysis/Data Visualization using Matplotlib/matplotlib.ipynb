```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## 2D plot
- Bivariate analysis
- used where one columns is numerical and other is categorical
- usecase: Times series data
- eg: increase in covid cases over the months Jan-Dec

```python
# plotting a simple funcion
cases = [58000,35000,22000,8055]
year = ['2020','2021','2022','2023']
plt.plot(year,cases)
```

```python
# plotting a dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Virat Kohli vs Rohit Sharma Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# color,linestyle,linewidth,marker,marker size
plt.plot(batsman['index'],batsman['V Kohli'],color = 'red',linestyle = 'dashed',linewidth = 3,marker = '>',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'blue',linestyle = 'dashed', linewidth = 3,marker = '>',markersize=10)
plt.title('Virat Kohli vs Rohit Sharma Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# legends -- > label attribute
plt.plot(batsman['index'],batsman['V Kohli'],color = 'red',linestyle = 'dashed',linewidth = 3,marker = '>',markersize=10,label = 'Virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'blue',linestyle = 'dashed', linewidth = 3,marker = '>',markersize=10, label= 'Rohit')
plt.title('Virat Kohli vs Rohit Sharma Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs')

plt.legend()
```

```python
# axes trimming
cases = [58000,35000,22000,234599]
year = ['2020','2021','2022','2023']
plt.plot(year,cases)
plt.ylim(0,100000)
plt.xlim('2021','2022')
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color = 'red',linestyle = 'dashed',linewidth = 3,marker = '>',markersize=10,label = 'Virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color = 'blue',linestyle = 'dashed', linewidth = 3,marker = '>',markersize=10, label= 'Rohit')
plt.title('Virat Kohli vs Rohit Sharma Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs')

plt.legend()
plt.grid()
```

## Scatter Plot
- Bivariate analysis
- Numerical vs Numerical
- Use case: Finding correlation

```python
# simple function
x = np.linspace(-10,10,50)
y = 10 * x + 3 + np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
# plotting a dataframe
# slower technique
df = pd.read_csv('batter.csv')
df = df.head(50)
plt.scatter(df['avg'],df['strike_rate'],color = 'red',marker = '.')
plt.title("Avg & SR of top 50 batsman")
plt.xlabel('Average')
plt.ylabel('Strike rate')           
plt.grid()
```

```python
# scatter plot using plot
# faster technique
plt.plot(df['avg'],df['strike_rate'],'o')
```

## Bar Chart
- Univariate or Bivariate
- Categorical vs Numerical
- Use case: aggregate analysis

```python
# simple function
votes = [20,40,5,6]
color = ['red','black','yellow','pink']
plt.bar(color,votes)
```

```python
# Horizontal bar
plt.barh(color,votes)
```

```python
df = pd.read_csv('batsman_season_record.csv')
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# overlapped categorical names
votes = [20,40,5,6]
color = ['red red red red','black black black black black','yellow yellow yellow yellow','pink pink pink pink']
plt.bar(color,votes)
```

```python
plt.bar(color,votes)
plt.xticks(rotation = 45)
plt.show()
```

```python
# Stacked bar chart
plt.bar(df['batsman'],df['2015'],label = '2015')
plt.bar(df['batsman'],df['2016'],bottom = df['2015'],label = '2016')
plt.bar(df['batsman'],df['2017'],bottom = (df['2015'] + df['2016']),label = '2017')
plt.legend()
```

## Pie chart
- Univariate/Bivariate analysis
- Categorical vs Numerical
- Use case: Find distribution on a standard scale

```python
# Simple function
marks = [56,35,23]
subjects = ['English','Hindi','Math']
plt.pie(marks,labels = subjects)

plt.show()
```

```python
# On dataframe
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels = df['batsman'],autopct = '%0.2f')

plt.show
```

```python
# Explode & shadow
plt.pie(df['batsman_runs'],labels = df['batsman'],autopct = '%0.2f',explode = [0.1,0,0,0,0,0],shadow = True)

plt.show
```
